```python
# ATEを推定
y = df['gamedummy']
z = df['cm_dummy']
e = df['e']
E_hat_y1 = ((z * y) / e).sum() / (z / e).sum()
E_hat_y0 = ((((1-z) * y) / (1-e)).sum()) / ((1-z) / (1-e)).sum()
E_hat_y1 - E_hat_y0
```
