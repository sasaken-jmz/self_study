# AVG・IPS・DR・SNIPS・SNDR推定量を勉強する
[参考: 反実仮想機械学習](https://gihyo.jp/book/2024/978-4-297-14029-8)
これら推定量の算出を行い、バイアス・バリアンスの計算を行う

## モジュールインポート

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from tqdm import tqdm
```

## データ生成

```python
random_seed = 12345
np.random.seed(random_seed)
n = 100_000

# 特徴量 X1〜X10
X = np.random.normal(0, 1, size=(n, 10))
columns = [f'X{i}' for i in range(1, 11)]

# r0: 非介入時（例えば X1〜X5 の線形結合 + ノイズ）
coeffs_r0 = np.random.uniform(-1, 1, size=5)
r0 = X[:, :5] @ coeffs_r0 + np.random.normal(0, 1, size=n)

# r1: 介入時（X6〜X10 の線形結合 + ノイズ + 介入効果）
coeffs_r1 = np.random.uniform(-1, 1, size=5)
intervention_effect = 2.0  # 介入効果
r1 = X[:, 5:] @ coeffs_r1 + np.random.normal(0, 1, size=n) + intervention_effect

# DataFrame作成
df = pd.DataFrame(X, columns=columns)
df['r0'] = r0
df['r1'] = r1

df.head()
```

```python
sample_n = 10_000
n_trials = 100
avg_estimator_list = []
dm_estimator_list = []
ips_estimator_list = []
snips_estimator_list = []
dr_estimator_list = []
sndr_estimator_list = []
true_r_list = []
for i in tqdm(range(n_trials)):
    random_seed = i
    np.random.seed(random_seed)
    
    # 1万行をランダム抽出
    df_sampled = df.sample(n=sample_n, random_state=random_seed).copy()
    
    # treatment割当（1: Tr / 0: Cr）
    df_sampled['pi_old_0'] = 0.5
    df_sampled['pi_old_1'] = 0.5
    df_sampled['treatment_flag'] = np.random.binomial(1, 0.5, size=sample_n)
    df_sampled['old_pscore'] = df_sampled['treatment_flag'] * df_sampled['pi_old_1'] + (1 - df_sampled['treatment_flag']) * df_sampled['pi_old_0']
    
    # treatmentに応じた観測報酬
    df_sampled['observed_reward'] = np.where(
        df_sampled['treatment_flag'] == 1,
        df_sampled['r1'],
        df_sampled['r0']
    )
    
    # treatment割当（1: Tr / 0: Cr）
    df_sampled['pi_new_0'] = 0.3
    df_sampled['pi_new_1'] = 0.7
    df_sampled['new_treatment_flag'] = np.random.binomial(1, 0.7, size=sample_n)
    df_sampled['new_pscore'] = df_sampled['treatment_flag'] * df_sampled['pi_new_1'] + (1 - df_sampled['treatment_flag']) * df_sampled['pi_new_0']
    
    # treatmentに応じた観測報酬
    df_sampled['new_observed_reward'] = np.where(
        df_sampled['new_treatment_flag'] == 1,
        df_sampled['r1'],
        df_sampled['r0']
    )

    # DR推定量用に予測値を用意する
    # 特徴量
    features = [f'X{i}' for i in range(1, 11)]
    
    # 処置フラグと報酬
    X = df_sampled[features]
    T = df_sampled['treatment_flag']
    Y = df_sampled['observed_reward']
    
    # モデル（treatment=0用）
    model_cr = RandomForestRegressor()
    model_cr.fit(X[T == 0], Y[T == 0])
    
    # モデル（treatment=1用）
    model_tr = RandomForestRegressor()
    model_tr.fit(X[T == 1], Y[T == 1])
    
    # 各個体について μ0, μ1 の予測値
    mu0_hat = model_cr.predict(X)
    mu1_hat = model_tr.predict(X)
    
    # データフレームに追加（後で DR に使える）
    df_sampled['mu0_hat'] = mu0_hat
    df_sampled['mu1_hat'] = mu1_hat
    df_sampled['mu_hat'] = T * mu1_hat + (1-T) * mu0_hat

    df_sampled['w'] = df_sampled['new_pscore'] / df_sampled['old_pscore']
    # AVG
    avg_estimator = df_sampled['observed_reward'].mean()
    avg_estimator_list.append(avg_estimator)
    # DM
    dm_estimator = (df_sampled['mu0_hat'] * df_sampled['pi_new_0'] + df_sampled['mu1_hat'] * df_sampled['pi_new_1']).mean()
    dm_estimator_list.append(dm_estimator)    
    # IPS
    ips_estimator = (df_sampled['w'] * df_sampled['observed_reward']).mean()
    ips_estimator_list.append(ips_estimator)
    # SNIPS
    snips_estimator = (1/df_sampled['w'].sum()) * (df_sampled['w'] * df_sampled['observed_reward']).sum()
    snips_estimator_list.append(snips_estimator)
    # DR
    dr_estimator = dm_estimator + (df_sampled['w'] * (df_sampled['observed_reward'] - df_sampled['mu_hat'])).mean()
    dr_estimator_list.append(dr_estimator)
    # SNDR
    sndr_estimator = dm_estimator + ((df_sampled['w']*(df_sampled['observed_reward']-df_sampled['mu_hat'])) / df_sampled['w'].mean()).mean()
    sndr_estimator_list.append(sndr_estimator)
    
    # 真の値
    true_r = df_sampled['new_observed_reward'].mean()
    true_r_list.append(true_r)
```

```python
def plot_estimator_diff(estimator_list, true_list, label="Estimator", bins=30):
    """
    推定値と真値の差分のヒストグラムを描き、誤差統計を表示する。

    Parameters:
        estimator_list (list or np.ndarray): 推定値のリスト
        true_list (list or np.ndarray): 真値のリスト
        label (str): 推定器の名前（グラフタイトル・出力に使用）
        bins (int): ヒストグラムのビン数
    """
    # NumPy配列に変換
    est = np.array(estimator_list)
    true = np.array(true_list)

    # 最小長に切り詰めて整合性確保
    min_len = min(len(est), len(true))
    est = est[:min_len]
    true = true[:min_len]

    # 差分計算 & 欠損値除去
    diff = est - true
    diff = diff[np.isfinite(diff)]

    # 誤差統計
    bias = np.mean(diff)
    variance = np.var(diff)
    mae = np.mean(np.abs(diff))
    mse = np.mean(diff**2)

    # ヒストグラム描画
    plt.figure(figsize=(8, 5))
    plt.hist(diff, bins=bins, edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1)
    plt.title(f"Difference Histogram: {label} - True")
    plt.xlabel("Difference")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

    # 統計表示
    print(f"[{label}]")
    print(f"  Bias (mean diff):      {bias:.5f}")
    print(f"  Variance (diff):       {variance:.5f}")
    print(f"  MAE (abs mean diff):   {mae:.5f}")
    print(f"  MSE (mean sq diff):    {mse:.5f}")
    print(f"  MSE ≈ Bias² + Var:     {bias**2 + variance:.5f}")
```

```python
plot_estimator_diff(avg_estimator_list, true_r_list, label="AVG")
plot_estimator_diff(dm_estimator_list, true_r_list, label="DM")
plot_estimator_diff(ips_estimator_list, true_r_list, label="IPS")
plot_estimator_diff(snips_estimator_list, true_r_list, label="SNIPS")
plot_estimator_diff(dr_estimator_list, true_r_list, label="DR")
plot_estimator_diff(sndr_estimator_list, true_r_list, label="SNDR")
```
